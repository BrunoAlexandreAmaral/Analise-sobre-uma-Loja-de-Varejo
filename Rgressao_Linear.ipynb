```python
df = pd.read_excel("df.xlsx")
```

```python
df.head()
```

```python
sns.lmplot(x="ano_2015", y="ano_2016", data = df);
```

```python
import statsmodels.api as sm
```

```python
y = df.ano_2016
x = sm.add_constant(df.ano_2015)
```

```python
resultado = sm.OLS(y,x).fit()
```

```python
print(resultado.summary())
```

```python
df["predicao"] = resultado.predict()
```

```python
df
```

```python
df["residuo"] = df["predicao"] - df["ano_2016"]
```
